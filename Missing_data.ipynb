```python
import pandas as pd

# Find Missing data in airtable as compared to spoors
print('Enter spoors Form ID data...')
a = [i for i in input().split(' ')]
b  = []
for x in a:
    try:
        b.append(int(x))
    except:
        pass

print('Enter Airtable Form ID data...')
c = [i for i in input().split(' ')]
d = []
for x in c:
    try:
        d.append(int(x))
    except:
        pass

set_difference = set(b) - set(d)
list_difference = list(set_difference)
print("OUTPUT:")
print(*list_difference)

md = pd.DataFrame(list_difference)
md.to_csv('missing_data.csv', index=False)

print("File created Download it")
```
